```python
# Function to take the first character of every answer(row).( Later converted into a combined score )
def score(value):
    temp = str(value)
    return temp[0]   
```

```python
# Function of List conversion to integer
def conv_to_int(lis):
    l = []
    for i in lis:
        a = int(i)
        l.append(a)
    return l
```

```python
no_rows = df2.shape[0]
df2['score'] = ''
df3 = df2.copy()
df3 = df3.drop(['Depression', 'score'], axis = 1)
df3.head()
```

```python
def score_mult(l):
    for i in range(no_rows):
        if i == 0 :
            l[i]*=4        
        elif i == 1 :
            l[i]*=2
        elif i == 2 :
            l[i]*=2
        elif i == 3 :
            l[i]*=3
        elif i == 4 :
            l[i]*=1
        elif i == 5 :
            l[i]*=1
        elif i == 6 :
            l[i]*=3
        elif i == 7 :
            l[i]*=2
        elif i == 8 :
            l[i]*=4
        elif i == 9 :
            l[i]*=3
        elif i == 10 :
            l[i]*=3
        elif i == 11 :
            l[i]*=4
        elif i == 12 :
            l[i]*=2
        elif i == 13 :
            l[i]*=2
        elif i == 14 :
            l[i]*=3
        elif i == 15 :
            l[i]*=4
        elif i == 16 :
            l[i]*=2
        elif i == 17 :
            l[i]*=1
        elif i == 18 :
            l[i]*2
        elif i == 19 :
            l[i]*=1
        elif i == 20 :
            l[i]*=4
    return l
li = [2, 2, 2, 4, 1, 1, 2, 4, 1, 4, 4, 4, 2, 1, 2, 4, 3, 2, 1, 2, 4]
li = score_mult(li)
```

```python
def score_calculator(dataframe):
    for i in range(no_rows):
        temp_row = df3.iloc[i] # based on df indexing this basically gives out rows
        temp_lis = list(temp_row.apply(score)) # passing each value of row 1
        lis = conv_to_int(temp_lis) # Converting str list to integer list
        lis = score_mult(lis)
        dataframe['score'][i] = reduce(lambda x,y : x+y, lis) # getting sum of all the values of a row i.e score and appending to the dataset
        # Update the Depression column
        if dataframe['score'][i] >= 135:
            dataframe['Depression'][i] = 1
        else:
            dataframe['Depression'][i] = 0
score_calculator(df2)
df2.head()
```
